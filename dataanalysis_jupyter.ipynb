```python
#Creating the bar chart 
plt.figure(figsize=(16,8)) 
sns.barplot(data=df,x='Category',y='Sales',palette='viridis') 
plt.title('Category wise sales') 
plt.xlabel('category') 
plt.ylabel('sales')
plt.legend()
plt.show()
```

```python
#Creating the report for total sales by the country 
plt.figure(figsize=(10,8)) 
sns.barplot(x='Country',y='Sales',data=df,hue='Region') 
plt.title('Total sales by country and region')
plt.xlabel('Country') 
plt.ylabel('Sales') 
plt.legend() 
plt.show()
```

```python
#Creating the vizualization for categoery - subcategory wise total sales 
plt.figure(figsize=(16,8)) 
sns.barplot(x='Category',y='Sales',data=df,hue='Sub-Category') 
plt.title('Total sales by category and sub category') 
plt.xlabel('category') 
plt.ylabel('sales') 
plt.legend() 
plt.show()
```

```python
df.columns
```

```python
k = df.groupby(['Category','Sub-Category'])[['Sales','Profit']].sum() 
k
```

```python
plt.figure(figsize=(10,8)) 
sns.barplot(data=k,x='Category',y='Sales',hue='Sub-Category') 
plt.title('Total sales and Total profits by category and subCATEGORY',fontweight='bold') 
plt.xlabel('category',fontweight='bold',fontweight = 'bold') 
plt.ylabel('sales',fontweight='bold') 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert date column
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Sort by date
df = df.sort_values('Order Date')

# Create line chart
plt.figure(figsize=(16,8))
sns.lineplot(data=df, x='Order Date', y='Sales', hue='Country')
plt.title('Line Chart for Year Analysis')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend(title='Country')
plt.show()
```
